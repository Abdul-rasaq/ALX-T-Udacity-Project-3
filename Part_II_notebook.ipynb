## Investigation Overview

 I want to know the performance and participation of the categorical variable like the member genders, user type, bike id ,bike shared during the trip and ride period. 


## Dataset Overview
There are 183412 rides in the dataset which consist of categorical variable majorly and few numerical variable like duration in seconds. 

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')
````

## (Visualization 1)

### Showing the sum of duration  by member gender in percentage

````python
df.groupby('member_gender').sum()['duration_sec'].plot(kind = 'pie', explode = [0.05,0.05,0.05],autopct = '%0.0f%%')
plt.title('Pie chart showing total duration by gender');
````

Male covered the highest duration with 71% follow by Female with 26 and the least, 3%. More are required to encourage to increase the participation of the females. 

## (Visualization 2)

### First ten Bikes that covered the highest duration 

````python
df.groupby('bike_id').sum()['duration_sec'].sort_values(ascending = False ).head(10).plot(kind = 'barh');
````

bike id 5481 covered the highest duration for the trip with more than 200000 sec. preceeding with bike id 5282 with roughly 200000 sec and 6235 covered the least duration among the 10. This information can be used to prepared maintenance check up for the bike. 

## (Visualization 3)

###  Showing the sum of duration by type of user in percentage


````python
df.groupby('user_type').sum()['duration_sec'].plot(kind = 'pie', explode = [0.05,0.05],autopct = '%0.0f%%')
plt.title('Pie chart showing total duration by user type');
````

Among the 2 user type, Total duration for the Subscribers trip is about 4 times of customer's trip. 

## (Visualization 4)

### showing the sum of duration by people that shared bike during trip in percentage


````python
df.groupby('bike_share_for_all_trip').sum()['duration_sec'].plot(kind = 'pie', explode = [0.05,0.05],autopct = '%0.0f%%')
plt.title('Pie chart showing total duration by people sharing bike during the trip');
````

Only few people shared bike during the trip with just 9% and 91% biker didn't shared bike during the trip.
